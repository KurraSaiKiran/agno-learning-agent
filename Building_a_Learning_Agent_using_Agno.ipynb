```python
# Imports
from agno.agent import Agent
from agno.models.openai import OpenAIChat
import os
```

```python
# Set your Groq API key
os.environ["GROQ_API_KEY"] = "ENTER_YOUR_API_KEY_HERE"
```

```python


# Create agent with Groq API (not Grok!)
agent = Agent(
    model=OpenAIChat(
        id="llama-3.3-70b-versatile",  # Groq's free model
        api_key=os.getenv("GROQ_API_KEY"),
        base_url="https://api.groq.com/openai/v1"  # Groq endpoint
    ),
    markdown=True,
)

print("✓ Agent created with Groq API!")
```

```python
# Test it
agent.print_response(
    "Hi! My name is Alice and I work as a research scientist at Anthropic.",
    stream=True,
)

agent.print_response(
    "I prefer concise responses without too much explanation.",
    stream=True,
)

agent.print_response(
    "What do you know about me?",
    stream=True,
)

agent.print_response(
    "I also love Python programming and machine learning.",
    stream=True,
)

agent.print_response(
    "Can you summarize everything you know about my interests?",
    stream=True,
)
```

```python
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

# Create input box and output area
input_box = widgets.Text(
    placeholder='Type your question here...',
    description='You:',
    layout=widgets.Layout(width='80%'),
    style={'description_width': 'initial'}
)

send_button = widgets.Button(
    description='Send',
    button_style='primary',
    layout=widgets.Layout(width='15%')
)

output_area = widgets.Output()

# Function to handle sending messages
def on_send(b):
    user_message = input_box.value
    if user_message.strip():
        with output_area:
            print(f"\n{'='*60}")
            print(f"You: {user_message}")
            print(f"{'='*60}\n")

            # Get agent response
            agent.print_response(user_message, stream=True)

        input_box.value = ''  # Clear input box

# Function to handle Enter key
def on_enter(text):
    on_send(None)

# Attach event handlers
send_button.on_click(on_send)
input_box.on_submit(on_enter)

# Display the chat interface
print("🤖 Chat Interface Ready! Type your questions below:\n")
display(widgets.HBox([input_box, send_button]))
display(output_area)
```
